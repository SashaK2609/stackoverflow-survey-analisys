```python
#Респонденти, які відповіли на обов*язкові питання
max_q = stackoverflow_df.dropna(subset = ['MainBranch', 'Age', 'Employment', 'Check', 'EdLevel', 'Country', 'Currency', 'NEWSOSites', 'AISelect', 'TBranch'])
print(len(max_q))
```

```python
#На скільки питань відповів кожен респондент
stackoverflow_df['count_q'] = stackoverflow_df.count(axis = 1).sort_values(ascending=False)
print(stackoverflow_df.head())
```

Серед респондентів, що програмують на  Python в групуванні по країнах, яка середня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні?

```python
grouped = python.groupby('Country')
comp_yearly = grouped['ConvertedCompYearly'].agg(['mean', 'median'])
print(comp_yearly)
```

В кожній віковій категорії який % респондентів програмує на Python?

```python
age = python.groupby('Age')
count_age = age['Age'].count()
print(count_age)
```

```python
#відсоток до загальної кількості респондентів
print(count_age / len(stackoverflow_df))
```

```python
#відсоток до кількості респондентів, які програмують на Python
print(count_age / len(python))
```

Серед респондентів що знаходяться в 75 перцентилі за компенсацією середнього і працюють віддалено, які індустрії є найрозповсюдженішими?

```python
q_75 = stackoverflow_df['ConvertedCompYearly'].quantile(q = 0.75)
filtered = stackoverflow_df[(stackoverflow_df['ConvertedCompYearly'] >= q_75) & (stackoverflow_df['RemoteWork'] == 'Remote')]
industry_counts = filtered['Industry'].value_counts()
print(industry_counts)
```
